# El problema de la multiplicación de una cadena de matrices

```python
def print_optimal_parens(s, i, j):
    if i == j:
        return f"A{i}"
    else:
        return f"({print_optimal_parens(s, i, s[i][j])} x {print_optimal_parens(s, s[i][j] + 1, j)})"

def matrix_chain_order(p):
    n = len(p) - 1
    m = [[0 for _ in range(n + 1)] for _ in range(n + 1)]
    s = [[0 for _ in range(n + 1)] for _ in range(n + 1)]

    for length in range(2, n + 1):
        for i in range(1, n - length + 2):
            j = i + length - 1
            m[i][j] = float('inf')
            for k in range(i, j):
                q = m[i][k] + m[k + 1][j] + p[i-1] * p[k] * p[j]
                if q < m[i][j]:
                    m[i][j] = q
                    s[i][j] = k

    return m, s

# Ejemplo de dimensiones: A1(15x20), A2(20x200), A3(200x11) y A4(11x300)
dimensions = [15, 20, 200, 11, 300]
m_table, s_table = matrix_chain_order(dimensions)
n = len(dimensions) - 1

print(f"Costo mínimo de multiplicaciones: {m_table[1][n]}")
print(f"Parentización óptima: {print_optimal_parens(s_table, 1, n)}")
```
