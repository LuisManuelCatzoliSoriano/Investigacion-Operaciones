```python
import numpy as np
from scipy.optimize import linprog
```

```python
# Costos (función objetivo)
c = [4,3,2,5,6,9,12,4,2,2,3,8]

# Matriz de restricciones (nutrientes)
A = [
    [-110,-90,-100,-150,-160,-200,-250,-180,-120,-50,-80,-200],  # calorias
    [-4,-3,-2,-8,-13,-20,-26,-12,-3,-1,0,-14],                   # proteina
    [-2,-1,0,-20,-5,-2,-3,-6,-2,-2,-1,-25],                      # calcio
    [-1,-1,-1,0,-2,-2,-3,-4,-1,-1,0,-1]                          # hierro
]

# Requerimientos mínimos
b = [-2000,-55,-800,-18]

# Restricciones de no negatividad
bounds = [(0,None)]*12

# Resolver el problema
resultado = linprog(c, A_ub=A, b_ub=b, bounds=bounds, method='highs')

# Mostrar resultados
print("Solución óptima:\n")

alimentos = [
"Avena","Pan","Arroz","Leche","Huevos","Pollo",
"Carne","Frijoles","Papas","Zanahorias","Manzanas","Queso"
]

for i in range(12):
    print(alimentos[i],":",resultado.x[i])

print("\nCosto mínimo de la dieta:",resultado.fun)
```
