#Problema de la mochila

### Explicación del Problema de la Mochila con Cantidades Finita (Entera)

En este ejercicio resolvemos el **problema de la mochila entera con cantidades limitadas**.  

## Datos del problema

- Tenemos $n$ objetos, cada uno con:
  - Volumen $w_i$  
  - Valor $v_i$  
  - Cantidad máxima disponible $c_i$  
- La capacidad/volumen máxima de la mochila es $W$.  

Queremos **maximizar el valor total** de los objetos que colocamos en la mochila, respetando:
1. La capacidad máxima $W$, es decir:
$
\sum_{i=1}^{n} x_i \cdot w_i \leq W
$
2. La cantidad disponible de cada objeto:
$
0 \leq x_i \leq c_i, \quad x_i \in \mathbb{Z}
$
donde $x_i$ representa cuántas unidades del objeto $i$ elegimos.

## Enfoque del código

- Usamos **programación dinámica** con una tabla `dp[j]` que representa el **valor máximo alcanzable** para una mochila de capacidad `j`.  
- Para manejar las cantidades limitadas, aplicamos la **descomposición binaria**, que convierte cada objeto con cantidad limitada en varios "objetos virtuales" de tipo 0-1.  
- Además, mantenemos una tabla `choice[j]` para **registrar cuántas unidades de cada objeto se eligen**, de modo que al final podamos reconstruir la combinación óptima.

Este método nos permite **no solo obtener el valor máximo**, sino también **saber exactamente qué y cuántos objetos tomar**.

## Datos del problema

Tenemos **10 objetos**, cada uno con un volumen, un valor y una cantidad máxima disponible:

| Objeto \(i\) | Volumen \(w_i\) | Valor \(v_i\) | Cantidad máxima \(c_i\) |
|---------------|----------------|---------------|-------------------------|
| 1             | 2              | 3             | 3                       |
| 2             | 3              | 4             | 2                       |
| 3             | 4              | 5             | 1                       |
| 4             | 5              | 6             | 2                       |
| 5             | 6              | 5             | 1                       |
| 6             | 7              | 10            | 2                       |
| 7             | 9              | 25            | 6                       |
| 8             | 5              | 11            | 1                       |
| 9             | 6              | 9             | 2                       |
| 10            | 2              | 7             | 7                       |

La **capacidad máxima de la mochila** es \( W = 17 \).

```python
# Problema de la mochila con cantidades finitas

# Datos de ejemplo
w = [2, 3, 4, 5, 6, 7, 9, 5, 6, 2]  # Volumen de cada objeto
v = [3, 4, 5, 6, 5, 10, 25, 11, 9, 7]  # Valor de cada objeto
c = [3, 2, 1, 2, 1, 2, 6, 1, 2, 7]  # Cantidad máxima disponible

n = len(w)  # número de objetos
W = 17  # Capacidad máxima de la mochila

# Inicializamos la DP y la tabla de selección
dp = [0] * (W + 1)
choice = [ [0]*n for _ in range(W + 1) ]  # choice[j][i] = cuántos del objeto i para capacidad j

# Iteramos sobre cada objeto
for i in range(n):
    cantidad = c[i]
    k = 1
    while cantidad > 0:
        take = min(k, cantidad)
        for j in range(W, take * w[i] - 1, -1):
            if dp[j] < dp[j - take * w[i]] + take * v[i]:
                dp[j] = dp[j - take * w[i]] + take * v[i]
                choice[j] = choice[j - take * w[i]].copy()
                choice[j][i] += take
        cantidad -= take
        k *= 2

print("Valor máximo que se puede obtener:", dp[W])
print("Cantidad de cada objeto elegida:", choice[W])
```
